# PSNR:
- Speak signal-to-noise ratio
- measure the quality of a reconstructed image or video when compared to the original, uncompressed version.
- Higher PSNR values (closer to infinity) imply a higher similarity and better quality between the original and compressed versions.
PSNR.png

# SSIM

- SSIM là viết tắt của "Structural Similarity Index" (Chỉ số Tương đồng Cấu trúc).
- Đây là một phép đo phổ biến được sử dụng để đo lường sự tương đồng giữa hai hình ảnh. SSIM đánh giá sự tương đồng cấu trúc giữa các hình ảnh, bao gồm độ sáng, độ tương phản và cấu trúc của chúng.

> Những điểm ảnh có mức độ sáng tối khác nhau, và càng có nhiều mức độ sáng tối càng có nhiều chi tiết ảnh => Ảnh chất lượng tốt


>   Một bức ảnh không phải độ tương phản càng cao thì càng tốt mà nên có sự hài hòa cân đối giữa sáng và tối. => Độ da dạng

- Đánh giá trên 2 ảnh xám và không cần cùng kích thước


```python
from skimage.metrics import structural_similarity as ssim
import cv2
import os
from google.colab.patches import cv2_imshow
import pandas as pd
```

## SATE1k_Thin

```python
excel_file_path = 'SATE1K_Thin_data(path).xlsx'
dir_target = '/content/drive/MyDrive/AIDTransformer/testing_data/Sate1K/Thin/target'
dir_result = '/content/drive/MyDrive/AIDTransformer/results/Dehazed/Sate1K_Thin'
```

```python
list_dir_target = sorted(os.listdir(dir_target))
list_dir_result = sorted(os.listdir(dir_result))
```

```python
print(len(list_dir_target))
print(len(list_dir_result))
```

ssim() return 2 values:
* ssim_index = SSIM score
* diff = shows the local SSIM values at different regions of the images
* full = 'True': tells the function to return the full structural similarity image (similarity map) in addition to the overall SSIM index.

```python
SSIM = []
for (target, result) in zip(list_dir_target, list_dir_result):
  img1 = cv2.imread(os.path.join(dir_target, target), 0)
  img2 = cv2.imread(os.path.join(dir_result, result), 0)
  #cv2_imshow(img1)
  #cv2_imshow(img2)
  resized_img1 = cv2.resize(img1, (256, 256))
  ssim_index, diff = ssim(resized_img1, img2, full = True)

  SSIM.append(ssim_index)
```

```python
len(SSIM)
```

```python
# Ghi DataFrame vào file Excel
df1 = pd.DataFrame(SSIM)
df1.to_excel(excel_file_path, index=True)
```

```python
%%shell
#!/bin/bash

image_directory="/content/drive/MyDrive/AIDTransformer/testing_data/RICE/input"
target_directory="/content/drive/MyDrive/AIDTransformer/testing_data/RICE/target"
AIDTransformer_directory="/content/drive/MyDrive/AIDTransformer/results/Dehazed/RICE"
output_file="/content/drive/MyDrive/AIDTransformer/RICE.html"

# Get the total number of files
total_files=$(find "${image_directory}" -type f | wc -l)
completed_files=0

# Create the HTML table
html=$(cat <<EOF
<!DOCTYPE html>
<html>
<head>
    <title>Image Table</title>
    <style>
        table {
            border-collapse: collapse;
        }
        th, td {
            border: 1px solid black;
            padding: 8px;
        }
        img {
            max-width: 100%;
            max-height: 100%;
        }
    </style>
</head>
<body>
    <table>
        <thead>
            <tr>
                <th>Image Path</th>
                <th>Original image</th>
                <th>Target</th>
                <th>Result</th>
            </tr>
        </thead>
        <tbody>
EOF
)

# Loop through the image files and add their Base64-encoded images to the HTML table
#index=1
for image_file in "${image_directory}"/*; do
    filename=$(basename "$image_file")
    original_image_path="${image_directory}/${filename}"
    annotation_image_path="${target_directory}/${filename}"
    cam_image_path="${AIDTransformer_directory}/${filename%.*}.png"
    #cam_image_path="${AIDTransformer_directory}/${filename}"
    #echo "1"

    # Read the image file and encode it in Base64 using the base64 command-line tool
    original_image_base64=$(base64 -w 0 "${original_image_path}")
    annotation_image_base64=$(base64 -w 0 "${annotation_image_path}")
    cam_image_base64=$(base64 -w 0 "${cam_image_path}")
    #echo "2"

    # Increment completed files count
    completed_files=$((completed_files + 1))
    #echo "3"

    # Print progress message
    echo "Processing file ${completed_files}/${total_files} ($(bc <<< "scale=2; ${completed_files} * 100 / ${total_files}"))%"
    #echo "4"

    # Add the row to the HTML table with embedded Base64 images
# for coco images with annotation
    html+=$(cat <<EOF
            <tr>
                <td>${original_image_path} </td>
                <td><img src='data:image/jpeg;base64,${original_image_base64}' alt="Original image"></td>
                <td><img src='data:image/jpeg;base64,${annotation_image_base64}' alt="Target"></td>
                <td><img src='data:image/jpeg;base64,${cam_image_base64}' alt="Result"></td>

            </tr>
EOF
)
  #index=$((index + 1))

# # for cub images without annotation
#     html+=$(cat <<EOF
#             <tr>
#                 <td><img src='data:image/jpeg;base64,${original_image_base64}' alt="Original Image"></td>
#                 <td><img src='data:image/jpeg;base64,${cam_image_base64}' alt="CAM Image"></td>
#             </tr>
# EOF
# )

done

# Close the HTML table
html+=$(cat <<EOF
        </tbody>
    </table>
</body>
</html>
EOF
)

# Save the HTML code to a file
echo "${html}" > "${output_file}"
```

# Change file name


```python
import os

# Directory containing the files to be renamed
directory_path = "/content/drive/MyDrive/AIDTransformer/testing_data/RICE/input"

# Get a list of files in the directory
file_list = os.listdir(directory_path)
file_list.sort()  # Sort the file list

# Iterate through sorted files and rename them with sequential numbers
for index, old_filename in enumerate(file_list, start=1):
    if os.path.isfile(os.path.join(directory_path, old_filename)):
        # Get the file extension
        file_extension = os.path.splitext(old_filename)[1]

        # Construct the new filename with sequential number
        new_filename = f"{index}{file_extension}"

        # Construct the full paths for old and new filenames
        old_filepath = os.path.join(directory_path, old_filename)
        new_filepath = os.path.join(directory_path, new_filename)

        # Rename the file
        os.rename(old_filepath, new_filepath)

        print(f"File '{old_filename}' renamed to '{new_filename}'")
```
